```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
from scipy.stats import linregress
from sklearn.ensemble import GradientBoostingRegressor
```

```python
%matplotlib inline
```

```python
tcs=pd.read_csv('tcs_updated.csv')
infy=pd.read_csv('infy_updated.csv')
```

```python
tcs=tcs.dropna(axis=0)
infy=infy.dropna(axis=0)
```

# LassoLars

```python
#Test Train Split
X_train, X_test, y_train, y_test = train_test_split(tcs['Close'], tcs['Close_next_day'], test_size=0.33, random_state=42)
X_train=X_train.values.reshape(-1,1)
X_test=X_test.values.reshape(-1,1)
```

```python
#LassoLars model imlimentatin
reg = linear_model.LassoLars(alpha=0.01)
reg=reg.fit(X_train, y_train)
pred=reg.predict(X_test)
print('R2 = ',r2_score(y_test, pred))
```

```python
# plot of observed vs predicted value
plt.scatter(y_test.values,pred,c='red', marker='o', alpha=1)
plt.title('observed vs predicted value')
slope, intercept, _,_,_ = linregress(pred,y_test.values)
plt.plot(pred, slope*pred + intercept,'-')
plt.show()
print('slope = ',slope,'intercept = ',intercept)
```

```python
#Test Train Split
X_train, X_test, y_train, y_test = train_test_split(infy['Close'], infy['Close_next_day'], test_size=0.33, random_state=42)
X_train=X_train.values.reshape(-1,1)
X_test=X_test.values.reshape(-1,1)
```

```python
#LassoLars model imlimentatin
reg = linear_model.LassoLars(alpha=0.01)
reg=reg.fit(X_train, y_train)
pred=reg.predict(X_test)
print('R2 = ',r2_score(y_test, pred))
```

```python
# plot of observed vs predicted value
plt.scatter(y_test.values,pred,c='red', marker='o', alpha=1)
plt.title('observed vs predicted value')
slope, intercept, _,_,_ = linregress(pred,y_test.values)
plt.plot(pred, slope*pred + intercept,'-')
print('slope = ',slope,'intercept = ',intercept)
plt.show()
```

## As R2 is above .85 in both the cases and slope of the line is almost 1 thus our LassoLars model gives acceptable prediction

# GradientBoostingRegressor

```python
#Test Train Split
X_train, X_test, y_train, y_test = train_test_split(tcs['Close'], tcs['Close_next_day'], test_size=0.33, random_state=42)
X_train=X_train.values.reshape(-1,1)
X_test=X_test.values.reshape(-1,1)
```

```python
#GradientBoostingRegressor model imlimentatin
reg = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1,max_depth=1, random_state=0, loss='ls')
reg=reg.fit(X_train, y_train)
pred=reg.predict(X_test)
print('R2 = ',r2_score(y_test, pred))
```

```python
# plot of observed vs predicted value
plt.scatter(y_test.values,pred,c='red', marker='o', alpha=1)
plt.title('observed vs predicted value')
slope, intercept, _,_,_ = linregress(pred,y_test.values)
plt.plot(pred, slope*pred + intercept,'-')
plt.show()
print('slope = ',slope,'intercept = ',intercept)
```
